```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

def load_data(file_path):
    """CSV 파일을 로드합니다."""
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"{file_path} 파일을 찾을 수 없습니다.")
    return pd.read_csv(file_path)

def describe_data(data):
    """데이터셋의 기본 통계 정보를 출력합니다."""
    print("\n[기술 통계]")
    print(data.describe(include='all'))

def plot_histograms(data):
    """각 열의 분포를 히스토그램으로 시각화합니다."""
    numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns

    for col in numeric_cols:
        plt.figure(figsize=(10, 6))
        sns.histplot(data[col], kde=True)
        plt.title(f'{col} 분포')
        plt.xlabel(col)
        plt.ylabel('빈도수')
        plt.grid(True)
        plt.show()

def plot_class_distribution(data, column_name):
    """특정 열의 클래스 분포를 시각화합니다."""
    plt.figure(figsize=(10, 6))
    sns.countplot(x=data[column_name])
    plt.title(f'{column_name} 클래스 분포')
    plt.xlabel(column_name)
    plt.ylabel('빈도수')
    plt.grid(True)
    plt.show()

def check_missing_values(data):
    """결측치가 있는 열을 확인합니다."""
    missing = data.isnull().sum()
    missing = missing[missing > 0]
    if not missing.empty:
        print("\n[결측치가 있는 열]")
        print(missing)
    else:
        print("\n결측치가 없습니다.")

def check_class_imbalance(data, target_column):
    """타겟 열의 클래스 불균형을 점검합니다."""
    class_counts = data[target_column].value_counts()
    print(f"\n[{target_column} 클래스 불균형 확인]")
    print(class_counts)

def check_correlation(data):
    """숫자형 열 간의 상관 관계를 시각화합니다."""
    corr = data.corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title('숫자형 열 간의 상관 관계')
    plt.show()
```

```python

def main():
    train_file = 'dataset/52_EventInfo_train.csv'
    val_file = 'dataset/52_EventInfo_val.csv'

    # 데이터 로드
    train_data = load_data(train_file)
    val_data = load_data(val_file)

    # 데이터 기술 통계 확인
    print("[Train Dataset]")
    describe_data(train_data)
    check_missing_values(train_data)

    print("\n[Validation Dataset]")
    describe_data(val_data)
    check_missing_values(val_data)

    # 히스토그램으로 분포 시각화
    print("\n[Train Dataset Histogram]")
    plot_histograms(train_data)

    print("\n[Validation Dataset Histogram]")
    plot_histograms(val_data)

    # 클래스 분포 확인
    print("\n[클래스 분포 확인]")
    plot_class_distribution(train_data, 'DAY_TYPE')
    plot_class_distribution(train_data, 'BUSROUTE_ID')
    plot_class_distribution(train_data, 'LOCATION_ID')

    # 클래스 불균형 점검
    check_class_imbalance(train_data, 'DAY_TYPE')

    # 숫자형 열 간 상관 관계 시각화
    print("\n[Train Dataset 상관 관계]")
    check_correlation(train_data)

    print("\n[Validation Dataset 상관 관계]")
    check_correlation(val_data)

if __name__ == "__main__":
    main()

```
